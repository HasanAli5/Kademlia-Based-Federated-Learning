```python
import pathlib
import matplotlib.pyplot as plt
import json

nodes = 10
```

```python
# kademlia based FL results
kbfl_logs = []

for i in range(nodes):
    node = pathlib.Path(f"./results/node_{i}").resolve()
    if node.exists():
        logs = node.glob("*.txt")
        latest_log = None
        for log in logs:
            if latest_log == None:
                latest_log = log
                continue
            current_timestamp = log.name.replace("log_","").replace(".txt","")
            lastest_timestamp = latest_log.name.replace("log_","").replace(".txt","")
            if float(current_timestamp)>float(lastest_timestamp):
                latest_log = log
        if latest_log:
            kbfl_logs.append(json.loads(latest_log.read_text()))

print(kbfl_logs)
```

```python
import numpy as np

numpy_logs = np.array(kbfl_logs)
kbfl_logs_average = np.mean(numpy_logs,axis=0).tolist()
```

```python


fl_logs = []

for i in range(nodes):
    node = pathlib.Path(f"./results_sim/node_{i}").resolve()
    if node.exists():
        logs = node.glob("*.txt")
        latest_log = None
        for log in logs:
            if latest_log == None:
                latest_log = log
                continue
            current_timestamp = log.name.replace("log_","").replace(".txt","")
            lastest_timestamp = latest_log.name.replace("log_","").replace(".txt","")
            if float(current_timestamp)>float(lastest_timestamp):
                latest_log = log
        if latest_log:
            fl_logs.append(json.loads(latest_log.read_text()))

print(fl_logs)
```

```python
import numpy as np

numpy_logs = np.array(fl_logs)
fl_logs_average = np.mean(numpy_logs,axis=0).tolist()
```

```python
cl_logs = []

node = pathlib.Path(f"./results_centralised").resolve()
logs = node.glob("*.txt")
for log in logs:
    cl_logs = json.loads(log.read_text())

print(cl_logs)
```

```python

from matplotlib.ticker import MultipleLocator
plt.figure(figsize=(10, 6))
plt.plot(kbfl_logs_average[0][0],label="KBFL Training Loss",color="blue",linestyle="--")
plt.plot(fl_logs_average[0][0],label="TFL Training Loss",color="orange",linestyle="--")
plt.plot(cl_logs[0][0],label="CL Training Loss",color="purple",linestyle="--")
plt.plot(kbfl_logs_average[1][0],label="KBFL Validation Loss",color="blue",linestyle="-",alpha=0.5)
plt.plot(fl_logs_average[1][0],label="TFL Validation Loss",color="orange",linestyle="-",alpha=0.5)
plt.plot(cl_logs[1][0],label="CL Validation Loss",color="purple",linestyle="-",alpha=0.5)

plt.legend(ncol=2, loc='upper center', frameon=True, fontsize='small')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid(True, linestyle='-', alpha=0.3)
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(5))
plt.xlim(left=-1,right=100)  # Starts the x-axis at 0
plt.ylim(bottom=0) # Starts the y-axis at 0
plt.title("Loss Comparison Between KBFL, TFL, and CL")
plt.show()
```
